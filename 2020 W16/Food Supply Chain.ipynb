```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://query.data.world/s/3d4y2xpmrawydtcrypjtyno6iuh4qs')
df
```

```python
df.describe()
```

## Add calculated columns and remove erroneous column

```python
#remove column
df = df.drop(['Unnamed: 8'], axis=1)

#update df with columns for pretransport and postprocessing with totals
df['PreTransport'] = df['Land use change']+df['Animal Feed']+df['Farm']+df['Processing']
df['PostProcessing'] = df['Transport']+df['Packging']+df['Retail']
```

```python
df
```

```python
df2 =pd.melt(df, id_vars='Food product', var_name='Supply chain area', value_name='Greenhouse gas emission per kg of food')
df2
```

```python
df3 = df2[(df2['Supply chain area'] == 'PreTransport') | (df2['Supply chain area'] == 'PostProcessing')]
df3
```

```python
df4 = df2[df2['Supply chain area'] == 'PreTransport'] 
df4
```

```python
PreTransport_emissions= df4['Greenhouse gas emission per kg of food'].sum()
PostProcessing_total_emissions = df3['Greenhouse gas emission per kg of food'].sum()
percent = PreTransport_emissions/PostProcessing_total_emissions
percent
```

## 91% of emissions occur before transport
